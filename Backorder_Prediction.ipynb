```python
# filling null values
# function for removing null values

def random_value_imputation(feature):
    random_sample = data[feature].dropna().sample(data[feature].isna().sum())
    random_sample.index = data[data[feature].isnull()].index
    data.loc[data[feature].isnull(), feature] = random_sample

random_value_imputation('sku')
random_value_imputation('national_inv')
random_value_imputation('lead_time')
random_value_imputation('in_transit_qty')
random_value_imputation('forecast_3_month')
random_value_imputation('forecast_6_month')
random_value_imputation('forecast_9_month')
random_value_imputation('sales_1_month')
random_value_imputation('sales_3_month')
random_value_imputation('sales_6_month')
random_value_imputation('sales_9_month')
random_value_imputation('min_bank')
random_value_imputation('potential_issue')
random_value_imputation('pieces_past_due')
random_value_imputation('perf_6_month_avg')
random_value_imputation('perf_12_month_avg')
random_value_imputation('local_bo_qty')
random_value_imputation('deck_risk')
random_value_imputation('oe_constraint')
random_value_imputation('ppap_risk')
random_value_imputation('stop_auto_buy')
random_value_imputation('rev_stop')
random_value_imputation('went_on_backorder')


# checking now the nullvalues
percentage1 = data.isnull().sum()
percentage2 = data.isnull().sum()/data.isnull().count()*100
missing_values=pd.concat([percentage1,percentage2], axis=1,keys=['Total','%'])
missing_values
```
